**Casting the `'Date'` column to a datetime type**

Use `pd.to_datetime` to convert the `Date` column to a datetime.

**WARNING**: The date column is formated as day-month-year. Does `pd.to_datetime` work the way you expect it to on day/month/year strings? **How many cases were there per year?**

Recall that with time series data, we can't simply group by on the year. You'll need to use the `resample` method on the DataFrame to answer this question.

```python
yearly = pox.resample('Y')

yearly
```

```python
yearly.agg(['sum', 'mean', 'min', 'max'])
```

### Plot rolling averages

Plot the weekly chickenpox cases, but on the same graph, also plot the 4-week and 12-week rolling averages.

```python
pox['4weeks_avg'] = pox["budapest_cases"].rolling(4).mean()
pox['12weeks_avg'] = pox["budapest_cases"].rolling(12).mean()
pox.dropna()
pox.head()

plt.figure(figsize=(15,10))
plt.plot(pox["budapest_cases"]);
plt.plot(pox["4weeks_avg"]);
plt.plot(pox["12weeks_avg"]);
```
